```python
def interpolation_search(arr, x):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high and x >= arr[low] and x <= arr[high]:
        probes += 1

        pos = low + ((x - arr[low]) * (high - low)) // (arr[high] - arr[low])

        if arr[pos] == x:
            return pos, probes
        elif arr[pos] < x:
            low = pos + 1
        else:
            high = pos - 1

    return -1, probes


def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high:
        probes += 1
        mid = (low + high) // 2

        if arr[mid] == x:
            return mid, probes
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1

    return -1, probes


arr = list(map(int, input("Enter sorted roll numbers: ").split()))
x = int(input("Enter roll number to search: "))

i_pos, i_probes = interpolation_search(arr, x)
b_pos, b_probes = binary_search(arr, x)

print("Interpolation Search Probes =", i_probes)
print("Binary Search Probes =", b_probes)

if i_pos != -1:
    print("Roll number found")
else:
    print("Roll number not found")
```
