```python
import math

INF = math.inf

graph = [
    [0, 4, 2, 0, 0, 0],
    [4, 0, 5,10, 0, 0],
    [2, 5, 0, 0, 3, 0],
    [0,10, 0, 0, 7, 8],
    [0, 0, 3, 7, 0, 6],
    [0, 0, 0, 8, 6, 0]
]

n = len(graph)

selected = [False] * n
selected[0] = True

print("Edges in MST:")

total = 0

for _ in range(n - 1):
    minimum = INF
    x = y = 0

    for i in range(n):
        if selected[i]:
            for j in range(n):
                if (not selected[j]) and graph[i][j]:
                    if graph[i][j] < minimum:
                        minimum = graph[i][j]
                        x = i
                        y = j

    print(chr(65 + x), "-", chr(65 + y), ":", graph[x][y])
    total += graph[x][y]
    selected[y] = True

print("Total Cost =", total)
```
