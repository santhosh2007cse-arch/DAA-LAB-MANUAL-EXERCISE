```python
import random

# Interpolation Search for Floating Point Numbers
def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and arr[low] <= key <= arr[high]:
        comparisons += 1

        # Avoid division by zero
        if arr[high] == arr[low]:
            if arr[low] == key:
                return low, comparisons
            return -1, comparisons

        pos = low + int(
            ((key - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        if pos < low or pos > high:
            break

        # Comparison
        if abs(arr[pos] - key) < 1e-9:
            return pos, comparisons

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    return -1, comparisons


# Dataset sizes
sizes = [10000, 50000, 100000]

for n in sizes:

    # Generate uniformly distributed floating-point numbers
    data = sorted(random.uniform(0.0, 1000.0) for _ in range(n))

    total_comparisons = 0
    num_trials = 100

    for _ in range(num_trials):
        # Select a random existing value
        key = random.choice(data)

        _, comparisons = interpolation_search(data, key)

        total_comparisons += comparisons

    average_comparisons = total_comparisons / num_trials

    print(f"\nDataset Size: {n}")
    print(f"Average Comparisons: {average_comparisons:.2f}")
```
