```python
def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high and key >= arr[low] and key <= arr[high]:
        probes += 1

        if arr[low] == arr[high]:
            if arr[low] == key:
                return low, probes
            break

        pos = low + ((key - arr[low]) * (high - low)) // (arr[high] - arr[low])

        if arr[pos] == key:
            return pos, probes
        elif arr[pos] < key:
            low = pos + 1
        else:
            high = pos - 1

    return -1, probes


def binary_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high:
        probes += 1
        mid = (low + high) // 2

        if arr[mid] == key:
            return mid, probes
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1

    return -1, probes


# Input sorted roll numbers
n = int(input("Enter number of students: "))

print("Enter sorted roll numbers:")
roll_numbers = list(map(int, input().split()))

key = int(input("Enter roll number to search: "))

# Interpolation Search
index_i, probes_i = interpolation_search(roll_numbers, key)

# Binary Search
index_b, probes_b = binary_search(roll_numbers, key)

print("\n--- Search Results ---")

if index_i != -1:
    print(f"Interpolation Search: Roll number found at index {index_i}")
else:
    print("Interpolation Search: Roll number not found")

print(f"Interpolation Search Probes: {probes_i}")

if index_b != -1:
    print(f"Binary Search: Roll number found at index {index_b}")
else:
    print("Binary Search: Roll number not found")

print(f"Binary Search Probes: {probes_b}")

# Comparison
print("\n--- Comparison ---")
if probes_i < probes_b:
    print("Interpolation Search required fewer probes.")
elif probes_i > probes_b:
    print("Binary Search required fewer probes.")
else:
    print("Both searches required the same number of probes.")
```
