```python
def interpolation_search(arr, key):
    low = 0
    high = len(arr) - 1
    probes = 0

    while low <= high and key >= arr[low] and key <= arr[high]:
        probes += 1

        if arr[low] == arr[high]:
            if arr[low] == key:
                return low, probes
            else:
                return -1, probes

        pos = low + int(
            ((key - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        if arr[pos] == key:
            return pos, probes

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    return -1, probes


# Example
arr = [10, 20, 30, 40, 50, 60, 70]
key = 50

index, probes = interpolation_search(arr, key)

if index != -1:
    print("Element found at index:", index)
else:
    print("Element not found")

print("Number of probes:", probes)
```

```python
import random


def interpolation_search_float(arr, key):
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and key >= arr[low] and key <= arr[high]:
        comparisons += 1

        if arr[low] == arr[high]:
            if arr[low] == key:
                return low, comparisons
            else:
                return -1, comparisons

        pos = low + int(
            ((key - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        # Ensure position stays within bounds
        pos = max(low, min(pos, high))

        if arr[pos] == key:
            return pos, comparisons

        elif arr[pos] < key:
            low = pos + 1

        else:
            high = pos - 1

    return -1, comparisons


# Dataset sizes to test
dataset_sizes = [10000, 50000, 100000]

print("Interpolation Search Analysis")
print("-" * 40)

for n in dataset_sizes:

    # Generate uniformly distributed floating-point numbers
    arr = sorted(random.uniform(0.0, 1000.0) for _ in range(n))

    # Choose an existing element so search succeeds
    key = arr[n // 2]

    index, comparisons = interpolation_search_float(arr, key)

    print(f"Dataset Size : {n}")
    print(f"Key Found At : {index}")
    print(f"Comparisons  : {comparisons}")
    print("-" * 40)
```
